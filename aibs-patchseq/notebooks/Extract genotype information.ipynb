```python
import pandas as pd
```

```python
#
# open the metadata file
#
cell_df = pd.read_csv("../downloaded/20200711_patchseq_metadata_mouse.csv")
```

```python
#
# create a donors table
#
donor_df = cell_df[['donor_id','donor_name', 'biological_sex','age','full_genotype']].copy()
donor_df.drop_duplicates(inplace=True)
donor_df.set_index('donor_id',inplace=True)
donor_df
```

```python
#
# write out donor table
#
donor_df.to_csv('../outputs/donors.csv')
```

```python
#
# create an allele and a donor_allele dataframe
#
allele_df = pd.DataFrame(columns=['allele_name'])
donor_allele_df = pd.DataFrame(columns=['donor_id','locus_index','allele_index','allele_name'])
```

```python
#
# For each donor, break genotype into locus and alleles
#

for row in donor_df.itertuples():
    
    print(row.donor_name)
    
    # split genotype into locus part
    locus_list = row.full_genotype.split(';')
    
    for locus_index in range(len(locus_list)) :
        
        locus = locus_list[locus_index]
        
        # split locus genotype into alleles
        allele_list = locus.split('/')
        #print((locus_index,allele_list))
        
        for allele_index in range(len(allele_list)) :
            
            allele = allele_list[allele_index]
            
            if not (allele in allele_df['allele_name'].values):
                allele_df = allele_df.append({'allele_name': allele}, ignore_index=True)
    
            donor_allele_df = donor_allele_df.append({'donor_id': int(row.Index),
                                                     'locus_index': int(locus_index),
                                                     'allele_index' : int(allele_index),
                                                     'allele_name' : allele}, ignore_index=True )
```

```python
donor_allele_df
```

```python
# write out allele table
allele_df.to_csv('../outputs/allele.csv', index=False)
donor_allele_df.to_csv('../outputs/donor_allele.csv', index=False)
```
