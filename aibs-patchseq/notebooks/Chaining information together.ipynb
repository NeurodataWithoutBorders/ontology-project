This notebook demostrate how to chain information together for the production of example files

Linkage chain
- File on DANDI
- Cell specimen
- Donor
- List of genotype (locus and allele)
- Locus
- Allele

```python
import pandas as pd
import numpy as np
```

```python
#
# open file manifest with links to nwb files
#
file_df = pd.read_csv('../downloaded/2020-07-08_mouse_file_manifest.csv')
```

```python
ind = file_df["technique"] == 'intracellular_electrophysiology'
nwb_df = file_df[ind]
nwb_df = nwb_df.astype({'cell_specimen_id': 'int32'})
nwb_df.head()
```

```python
#
# use first row as example
#
ex_file_row = nwb_df.iloc[0]
ex_file_row
```

```python
#
# open the cell metadata spreadsheet
#
cell_df = pd.read_csv('../downloaded/20200711_patchseq_metadata_mouse.csv')
cell_df.set_index('cell_specimen_id', inplace=True)
cell_df.head()
```

```python
#
# find the cell associated with the example file
#
ex_cell_row = cell_df.loc[ex_file_row['cell_specimen_id']]
ex_cell_row
```

```python
#
# open the donor spreadsheet
#
donor_df = pd.read_csv('../outputs/donors.csv')
donor_df.set_index('donor_id', inplace=True)
donor_df.head()
```

```python
#
# find the donor associated with the example file
#
ex_donor_row = donor_df.loc[ex_cell_row['donor_id']]
ex_donor_row
```

```python
#
# open the donor to allele spreadsheet
#
donor_allele_df = pd.read_csv('../outputs/donor_allele.csv')
donor_allele_df.head()
```

```python
#
# select the rows associate with the example donor
#
ind = donor_allele_df['donor_id'] == ex_cell_row['donor_id']
ex_donor_allele = donor_allele_df[ind]
ex_donor_allele
```

```python
#
#  open the allele information excel spreadsheet
#
allele_df = pd.read_excel('../curated/allele_resources.xlsx')
allele_df.set_index('allele_symbol', inplace=True)
allele_df.head()
```

```python
#
# for the example donor, find the number of genotype/locus
#
locus_list = np.unique(ex_donor_allele['locus_index'].values)
```

```python
#
# select the first locus as example
#
ex_locus_index = locus_list[0]
ex_locus_index
```

```python
#
# for the example locus, get information for all the alleles
#
ind = ex_donor_allele['locus_index'] == ex_locus_index
ex_alleles = ex_donor_allele[ind]
ex_alleles
```

```python
#
# select the first allele as example
#
ex_allele_symbol = ex_alleles.iloc[0]['allele_symbol']
ex_allele_symbol
```

```python
#
# get details about the allele
#
ex_allele_details = allele_df.loc[ex_allele_symbol]
ex_allele_details
```
